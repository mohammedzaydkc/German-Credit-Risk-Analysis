# Preprocess the dataset:
**We have to convert all categorical values into numerical values.**

#### Convert purpose column using Dummy variables

```python
dff=pd.get_dummies(df['Purpose'],drop_first=True,dtype=int)
```

```python
df=pd.concat([df.drop('Purpose',axis=1),dff],axis=1)
```

```python
df.head()
```

#### Use label encoding for rest of the categorical data

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in df.select_dtypes('O').columns:
    df[col] = le.fit_transform(df[col])
    
```

```python
df.head()
```

### Check the unique values in the new dataframe

```python
for i in df.columns:
    print(f'Unique values in new df coloumn {i} are :{df[i].unique()}')
```

```python
sns.heatmap(df.corr(),annot=True)
```

## Select the Dependent(target) and Independent Variables:

```python
X,y=df.drop('Risk',axis=1),df['Risk']
```

# Standardize the data

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
```

```python
X=pd.DataFrame(st.fit_transform(X),columns=st.get_feature_names_out())
X.head()
```

# Split the data into training and testing set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=100)
```

# Models:

## Random Forest Classifier.
** Let's choose the best estimator and parameters :GridSearchCV**

```python
#let's check what params will be best suitable for random forest classification.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
rnn=RandomForestClassifier(n_estimators=44)
rnn.fit(X_train,y_train)
```

```python
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 5, 10],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}
```

```python

grid=GridSearchCV(rnn,param_grid,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
print('best parameter',grid.best_params_)
print('best estimator',grid.best_estimator_)
best_rfc=grid.best_estimator_
y_pred=best_rfc.predict(X_test)
y_pred
```

### Check the accuracy score, Confusion metrics, Classification report and Cross val score

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print('accuracy score:',accuracy)
print('*********************************************************')
print('Classification report:',classification_report(y_test,y_pred))
print('*********************************************************')
print('confusion metrics:',confusion_matrix(y_test,y_pred))
print('*********************************************************')
cv_scores = cross_val_score(grid.best_estimator_, X_train, y_train, cv=5, scoring='accuracy')
print('cross val score:',cv_scores)
print('*********************************************************')
print('mean cross val score:',cv_scores.mean())
```

#### Visualize the confusion matrix

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
```

## PCA to Speed up Machine Learning Algorithms (Logistic Regression)
Step 0: Import and use PCA. After PCA you will apply a machine learning algorithm of your choice to the transformed data

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
pca.n_components
pca.fit(X_train)
```

```python
#Apply the mapping (transform) to both the training set and the test set.
train_X = pca.transform(X_train)
test_X = pca.transform(X_test)
```

**Step 1: Import the model you want to use**

In sklearn, all machine learning models are implemented as Python classes

```python
from sklearn.linear_model import LogisticRegression
```


**Step 2: Make an instance of the Model**

```python
# all parameters not specified are set to their defaults
# default solver is incredibly slow thats why we change it
# solver = 'lbfgs'
logisticRegr = LogisticRegression(solver = 'lbfgs')
```

```python
logisticRegr.fit(train_X,y_train)
```

```python
logisticRegr.score(train_X,y_train)
```

```python
y_pred=logisticRegr.predict(test_X)
```

```python
#Confusion Matrix
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
print('confusion metrics',metrics.confusion_matrix(y_test,y_pred))
```

## Model Deployment With Tkinter

### Import Tkinter Library

```python
import tkinter as tk
from tkinter import ttk
import joblib
#from PIL import ImageTk,Image

class model:
    def __init__(self, root):
        self.root = root
        self.root.title("Model Deployement with tkinter")

        # Create a main frame to hold all the widgets
        main_frame = tk.Frame(root)
        main_frame.pack(padx=10, pady=10)

        #images=tk.PhotoImage(file="chop.png")

        #bg_label=tk.Label(main_frame,image=images)
       # bg_label.grid(row=0,column=0,rowspan=10,columnspan=3,sticky='nsew')
        
        row = 0 

        # Age
        tk.Label(main_frame, text="Enter Age:").grid(row=row, column=0, padx=5,  pady=(5, 10))
        self.age_entry = tk.Entry(main_frame, width=20)
        self.age_entry.grid(row=row, column=1, padx=5, pady=5)
        row += 1
        

        # Sex
        tk.Label(main_frame, text="Choose Sex:").grid(row=row, column=0,padx=5, pady=(5, 10))
        self.sex_var = tk.StringVar()
        self.sex_var.set("Male")
        self.male_radio = tk.Radiobutton(main_frame, text="Male", variable=self.sex_var, value="Male")
        self.male_radio.grid(row=row, column=1,padx=5, pady=5)
        self.female_radio = tk.Radiobutton(main_frame, text="Female", variable=self.sex_var, value="Female")
        self.female_radio.grid(row=row, column=2,padx=5, pady=5)
        row += 1

        # Job
        tk.Label(main_frame, text="Choose Job Level:").grid(row=row, column=0, padx=5, pady=(5, 10))
        self.job_var = tk.StringVar()
        self.job_combobox = ttk.Combobox(main_frame, textvariable=self.job_var)
        self.job_combobox['values'] = ('Unskilled and non-resident', 'Skilled and non-resident', 'Skilled and resident', 'Manager')
        self.job_combobox.current(0)
        self.job_combobox.grid(row=row, column=1, padx=5, pady=5)
        row += 1

        # Housing
        tk.Label(main_frame, text="Choose Housing:").grid(row=row, column=0, padx=5, pady=(5, 10))
        self.house_var = tk.StringVar()
        self.house_combobox = ttk.Combobox(main_frame, textvariable=self.house_var)
        self.house_combobox['values'] = ('ddd', 'fffff', 'fffffff', 'fffffffffa')
        self.house_combobox.current(0)
        self.house_combobox.grid(row=row, column=1, padx=5, pady=5)
        row += 1

        # Saving
        tk.Label(main_frame, text="Choose Saving:").grid(row=row, column=0, padx=5,pady=(5, 10))
        self.saving_var = tk.StringVar()
        self.saving_combobox = ttk.Combobox(main_frame, textvariable=self.saving_var)
        self.saving_combobox['values'] = ('ddd', 'fffff', 'fffffff', 'fffffffffa')
        self.saving_combobox.current(0)
        self.saving_combobox.grid(row=row, column=1, padx=5, pady=5)
        row += 1

        # Checking
        tk.Label(main_frame, text="Choose Checking:").grid(row=row, column=0, padx=5, pady=(5, 10))
        self.checking_var = tk.StringVar()
        self.checking_combobox = ttk.Combobox(main_frame, textvariable=self.checking_var)
        self.checking_combobox['values'] = ('ddd', 'fffff', 'fffffff', 'fffffffffa')
        self.checking_combobox.current(0)
        self.checking_combobox.grid(row=row, column=1, padx=5, pady=5)
        row += 1

        # Credit Amount
        tk.Label(main_frame, text="Enter Credit Amount:").grid(row=row, column=0, padx=5,pady=(5, 10))
        self.credit_entry = tk.Entry(main_frame, width=20)
        self.credit_entry.insert(0, "Enter Credit Amount")
        self.credit_entry.grid(row=row, column=1, padx=5, pady=5)
        
        row += 1

        # Duration
        tk.Label(main_frame, text="Enter Duration:").grid(row=row, column=0, padx=5,pady=(5, 10))
        self.duration_entry = tk.Entry(main_frame, width=20)
        self.duration_entry.insert(0, "Enter Duration")
        self.duration_entry.grid(row=row, column=1, padx=5, pady=5)


        
        label = tk.Label(root, text="Prediction Output:")
        label.pack()

        output_entry = tk.Entry(root, width=40)
        output_entry.pack()

        button = tk.Button(root, text="Predict", command=predict)
        button.pack()

root = tk.Tk()
app = model(root) 
root.mainloop()

joblib.dump(logisticRegr,'model.pkl')
self.model=joblib.load('model.pkl')
```

### Define a function that will return the output of our prediction in an Entry box in tkinter window

### Create Tkinter window and all necessary input and output boxes

```python
def predict():
    age=float(self.age_entry.get())
    sex=self.sex_var.get()
    job=self.job_var.get() 
    housing=self.house_var.get() 
    saving=self.saving_var.get()
    checking=self.checking_var.get()
    credit_ammount=float(self.credit_entry.get()) 
    duration=float(self.duration_entry.get())
    output = model.predict([[age, sex, job, housing, saving, checking, credit_amount, duration]])

    # Display the output in the output field
    output_entry.delete(0, tk.END)
    output_entry.insert(0, str(output[0]))
```
